```python
import altair as alt
from vega_datasets import data
import pandas as pd
```

```python
df = pd.read_csv("data_with_id.csv",index_col =0)
df.head()
```

```python
df['Indicator'].value_counts()
```

```python
seats_parliament = df.loc[df.Indicator == "EMP17",:]
seats_parliament
```

```python
board_members = df.loc[df.Indicator == "EMP11",:]
board_members
```

```python
#currently using 2019 data with 3 countries
countries_1 = board_members.Country.tolist()
year =2019
filter_country = board_members[board_members.Country.isin(countries_1)]
filter_year = filter_country[filter_country.Year == year]
#filter_year.rename( columns={'Unnamed: 6':'id'}, inplace=True )
filter_year.to_csv('filter_year.csv')
```

```python
filter_year
```

```python
countries_2 = board_members.Country.tolist()
year =2019
filter_country2 = seats_parliament[seats_parliament.Country.isin(countries_2)]
filter_year2 = filter_country2[filter_country2.Year == year]
#filter_year.rename( columns={'Unnamed: 6':'id'}, inplace=True )
filter_year2.to_csv('filter_year2.csv')
```

```python
filter_year2
```

```python
countries = alt.topo_feature(data.world_110m.url, 'countries')
click = alt.selection_multi(fields=['Country'])
chart = (alt.Chart(countries).mark_geoshape()
     .transform_lookup(
         lookup='id',
         from_=alt.LookupData(filter_year, 'id', ['Value','Country'])
     )
     .encode(
         color='Value:Q',
         opacity=alt.condition(click, alt.value(1), alt.value(0.5)),
         tooltip=['Country:N', 'Value:Q']
     )
     .add_selection(click).project(
         type='mercator',
         scale=355,
         center=[-40,65],
         clipExtent= [[0, 0], [1000, 400]]
     )
     .properties(
         title='Europe Female Share of Seats for 2019',
         width=500,
         height=300
     )
)

bars = (
    alt.Chart(
        filter_year.nlargest(25, 'Value'),
        title='Female Share of Seats in National Parliaments').mark_bar().encode(
    x='Value',
    opacity=alt.condition(click, alt.value(1), alt.value(0.5)),
    color='Value',
    y=alt.Y('Country', sort='x'))
.add_selection(click))

chart & bars



```
